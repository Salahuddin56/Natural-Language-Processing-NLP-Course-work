**Bigram**

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import bigrams
from collections import Counter, defaultdict

def calculate_bigram_probability(corpus, text):
    # Tokenize the corpus and the text
    corpus_tokens = nltk.word_tokenize(corpus)
    text_tokens = nltk.word_tokenize(text)

    # Create bigrams
    corpus_bigrams = list(nltk.bigrams(corpus_tokens))
    text_bigrams = list(nltk.bigrams(text_tokens))

    # Frequency distribution of bigrams in the corpus
    corpus_bigram_freq = nltk.FreqDist(corpus_bigrams)

    # Frequency distribution of unigrams in the corpus
    corpus_unigram_freq = nltk.FreqDist(corpus_tokens)

    # Calculate bigram probability
    bigram_probability = {bigram: corpus_bigram_freq[bigram] / corpus_unigram_freq[bigram[0]] if bigram in corpus_bigram_freq else 0 for bigram in text_bigrams}

    return bigram_probability

# Read the corpus
with open('Bigram_Dataset.txt', 'r') as f:
    corpus = f.read()

# User input text
text = input("Enter your text: ")

# Calculate bigram probability
bigram_probability = calculate_bigram_probability(corpus, text)

# Print bigram probabilities
for bigram, prob in bigram_probability.items():
    print(f"Bigram: {bigram}, Probability: {prob}")
```
