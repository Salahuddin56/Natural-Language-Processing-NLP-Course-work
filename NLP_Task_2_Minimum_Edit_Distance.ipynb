**Minimum Edit Distance**

```python
import nltk
from nltk.metrics.distance import edit_distance

def get_closest_words(word, corpus_file='Minimum_Edit_Distance_Dataset.txt'):
    with open(corpus_file, 'r') as f:
        corpus = f.read().split()

    closest_words = [w for w in corpus if edit_distance(word, w) == 1]
    return closest_words

# Example usage:
word = input("Enter a word: ")
print(get_closest_words(word))
```
