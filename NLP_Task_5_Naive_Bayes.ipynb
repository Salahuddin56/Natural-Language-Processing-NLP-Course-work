**Naive Bayes**

```python
import pandas as pd

# Define data
data = {
    'labels': ['positive', 'negative', 'positive', 'negative', 'positive', 'negative', 'positive', 'negative', 'positive', 'negative','positive','positive'],
    'training': ['The product is very good and helpful',
                 'The service of the company is very bad',
                 'The person is educated and kind',
                 'He is educated but a liar',
                 'I love this product',
                 'I hate this service',
                 'This is a great movie',
                 'This is a terrible movie',
                 'The food was delicious',
                 'The food was awful',
                 'The weather was nice',
                 'This is a nice apartment']
}

# Create DataFrame
df = pd.DataFrame(data)

# Write DataFrame to CSV
df.to_csv('Naive_Bayes_Dataset.csv', index=False)
```

```python
def process_data(csv_file_path, test_data):
    df = pd.read_csv(csv_file_path)

    p_c = {}
    cls = set(df.labels)  # find the unique labels
    for c in cls:
      p_c[c] = len([1 for l in df.labels if l==c])/len(df.labels)
    # p_c

    all_corpus = ''
    pos_corpus = ''
    neg_corpus = ''
    for i in range(len(df.training)):
      if df.labels[i] == 'positive':
        pos_corpus = pos_corpus + ' '+ df.training[i]
      else:
        neg_corpus = neg_corpus + ' '+ df.training[i]
    all_corpus = pos_corpus + ' ' + neg_corpus

    all_vocab_count = len(set(all_corpus.split()))
    pos_corpus_count =  len(pos_corpus.split())
    neg_corpus_count =  len(neg_corpus.split())

    p = p_c
    test_result = {}
    all_vocab = set(all_corpus.split())   # all unique vocabulary

    for c in range(len(test_data)):
      test_words = test_data[c].split()
      # print(words)
      p_test = p.copy()

      for test_word in test_words:
        if test_word in all_vocab:
          p_test['positive'] = p_test['positive']*(pos_corpus.count(test_word)+1)/(pos_corpus_count+all_vocab_count)
          p_test['negative'] = p_test['negative']*(neg_corpus.count(test_word)+1)/(neg_corpus_count+all_vocab_count)
      # print(p)
      if p_test['positive']>p_test['negative']:
        test_result[test_data[c]] = ('positive', p_test['positive'])
      elif p_test['positive']<p_test['negative']:
        test_result[test_data[c]] = ('negative', p_test['negative'])
      else:
        test_result[test_data[c]] = ('failed!', 0.5)
    return(test_result)
```

```python
train_data_path = 'Naive_Bayes_Dataset.csv'
test_data = ['He is a bad boy',
             'She is good singer',
             'it was a nice car',
             'it was terrible accident',
             'The Mid Term Exam was easy']

result = process_data(train_data_path, test_data)

for key, value in result.items():
  print(f"{key} :: {value[0]}, Probability: {value[1]}")
```
